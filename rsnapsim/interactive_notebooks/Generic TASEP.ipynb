```python
#setup, lets load a large gene (HUMINSR) ~1700bp
import numpy as np
import os

os.chdir('..')

from rss import rSNAPsim
from rss import ProbeVectorFactory as pvf
from rss import PropensityFactory as pff
from rss import TranslationSolvers as tss
import matplotlib.pyplot as plt
import time
from rss import IntensityAnalyses as ia



plt.style.use('dark_background')
plt.rcParams['figure.dpi'] = 120
plt.rcParams['lines.linewidth'] = 1
plt.rcParams['axes.linewidth'] = 1.5
plt.rcParams['font.size'] = 12
plt.rcParams['axes.grid'] = False




os.chdir('./interactive_notebooks')

rsim = rSNAPsim()
rsim.open_seq_file('../gene_files/HUMINSR.gb')
```

```python
from rss import ModelBuilder as mb
poi = rsim.proteins['1'][0]  #protein object
model = mb()
model.add_k(poi)
model._poi.generate_3frame_tags()
```

```python
model.visualize_transcript()
```

```python
model.add_enters([5], [0], [.08])
model.add_jumps([200,150],[0,0], [200,350], [1,0], [3,3])
model._poi.multiframe_epitopes[1]['T_New'] = [ 240,250,230,260,270]
model.visualize_transcript()
```

```python
t = np.linspace(0,1000,1001)

ssa_soln = model.run_ssa(t,n_traj=50)
```

```python
plt.plot(ssa_soln[:,1,:].T, '#1cfff7',alpha=.3)
plt.plot(ssa_soln[:,0,:].T, '#08ff08',alpha=.3)
plt.xlabel('Time')
plt.ylabel('Intensity')
```
