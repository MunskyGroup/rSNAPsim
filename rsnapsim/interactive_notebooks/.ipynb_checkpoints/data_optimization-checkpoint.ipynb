## Calculating Autocovariance

```python
def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result[result.size // 2:]
```

```python
# Calculating the autocovariance
n_lags = 600
exp_int_dist = []
ac_array = np.zeros((total_n_spots , n_lags ))
norm_ac_array = np.zeros((total_n_spots , n_lags ))
counter = 0
# Calculating Autocorrelation.
for i in range(0,n_cells):
        for j in range(0,max_spots):
            temp_len_spot =  df[(df['Cell_#'] == i) & (df['Spot_#'] ==j) ]
            if len(temp_len_spot) >0:
                time = df[(df['Cell_#'] ==i) & (df['Spot_#'] ==j)]['Time_(sec)']
                intensity_red = df[(df['Cell_#'] ==i) & (df['Spot_#'] ==j)]['R_GaussianInt']
                intensity_green = df[(df['Cell_#'] ==i) & (df['Spot_#'] ==j)]['G_GaussianInt']
                intensity_blue = df[(df['Cell_#'] ==i) & (df['Spot_#'] ==j)]['B_GaussianInt']                
                exp_int_dist.append(df[(df['Cell_#'] ==i) & (df['Spot_#'] ==j)]['G_GaussianInt'].values[0]) # intensity values at the beggining
                
                temp_ac = autocorr(intensity_green)
                size_ac_temp  = len(temp_ac)
                ac_array[counter, 0:size_ac_temp] = autocorr(intensity_green)
                norm_ac_array[counter, 0:size_ac_temp] = autocorr(intensity_green)/ float(ac_array[counter, :] .max())
                plt.plot(ac_array[counter, :]  / float(ac_array[counter, :] .max()),'-' + 'g')
                counter += 1
```

```python
# Plotting mean autocovariance
lag_time = [*range(0, 100, 1)]
lag_time = [element * sampling_rate for element in lag_time]
mean_ac_data = norm_ac_array.mean(0)
std_ac_data = norm_ac_array.std(0)

# normalized autocovariance, removing shot noise
mean_ac_data_norm = mean_ac_data[1:-1]/mean_ac_data[1]

plt.plot(lag_time[1:100], mean_ac_data[1:100] , 'og');
plt.fill_between(lag_time[1:100], mean_ac_data[1:100] - std_ac_data[1:100], mean_ac_data[1:100] + std_ac_data[1:100], color='gray', alpha=0.2)

plt.xlim([1, 500]); # Removing the shot noise
plt.ylim([-0.1, 1]); 
```

# 3. - Comparing Experimental Data and Model Simulations.

## Comparing Autocovariance

Likelihood function to compare the autocovariance function for the fluorescence intensity predicted by the model to that of the measured time-lapse experiments.

\begin{equation*}
\log L(G|M) = C_1 - \frac{1}{Nt} \sum_{i=1}^{N_{t}} \frac{(G_{D}(\tau_{i}) - G_{M}(\tau_{i}))^2}{\sigma_{\tilde{G}_{D}(\tau_{i})}},
\end{equation*}

where $G_D$ and $G_M$ represent the experimental and simulated autocorrelation functions, respectively, and $\sigma_{\tilde{x}}$ is the standard error of the mean.

```python
objetive_fun_ac = - (total_n_spots/2) * (sum( (( mean_ac_data [2:100] - ssa_traj.mean_autocorr_norm[2:100])**2)/ std_ac_data[2:100] ) )
objetive_fun_ac
```

## Comparing Distributions

Distance between the simulated and measured distributions.

\begin{equation*}
    \log L(Dist|M) =  C_1 + \sum_{j=1}^{N_{s}} log P(j|M),
\end{equation*}

\begin{equation*}
    \log L(Dist|M) =  C_1 + \sum_{j=1}^{N_{b}} n_j log P(j|M),
\end{equation*}

where $N_{s}$ represents the number of experimental spots, $C_1$ is a constant, and $ P(j|M)$ is  the probability for an mRNA to have a fluorescence intensity corresponding to $j$ units of mature proteins (ump) given the model. As non-translating spots could not be separated from spots below a basal FLAG intensity in the experimental data measurements, comparison between simulations and measured distributions ignore all spots with an intensity value less than $1/2$ ump.

```python
dist_sim_data[dist_sim_data==0]= 1e-4 ;
objetive_fun_dist = -  np.dot(hist_exp_data,np.log(dist_sim_data))
objetive_fun_dist
```

## Comparing Spot Classification

Likelihood to observe the experimentally determined number of spots classification.

\begin{equation*}
    \log L(Dist|M) =  C_2 + \sum_{j=1}^{N_{D}} log P(f_j;\Lambda),
\end{equation*}

where each $f_j$ corresponds to one of the four spot types (i.e., Cap, IRES, both, or non-translating), $N_D$ is the number of independent observed spots, and $P\left(f_j;\Lambda\right)$ is the categorical distribution of spots of each type estimated by the model simulations with parameters $\Lambda$.

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e
```

```python
bounds = [(-5, 5), (-5, 5)]
```

```python
result = differential_evolution(ackley, bounds)
```
